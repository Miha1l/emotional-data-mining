## Построение гистограмм признаков

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./crowd_small_mean.csv')
data
```

```python
positive = data[data['emotion'] == 'positive'].iloc[:, 0:64]
sad = data[data['emotion'] == 'sad'].iloc[:, 0:64]
neutral = data[data['emotion'] == 'neutral'].iloc[:, 0:64]
angry = data[data['emotion'] == 'angry'].iloc[:, 0:64]
```

```python
def plot_features_hist(values, title):
    plt.bar([i for i in range(0, 64)], values)
    plt.title(title)
    plt.ylim(0, -80)
```

```python
plot_features_hist(positive.mean(axis=0).values, "Positive")
```

```python
plot_features_hist(sad.mean(axis=0).values, "Sad")
```

```python
plot_features_hist(angry.mean(axis=0).values, "Angry")
```

```python
plot_features_hist(neutral.mean(axis=0).values, "Neutral")
```
