```python
data_path = "C:\Study\emotional_data"

positive_sample_path = f"{data_path}/features/6a55ec57bc205ec887af4f59134d286e.npy"
sad_sample_path = f"{data_path}/features/4cb3546804898bd6760acce6e47a0938.npy"

positive_sample = np.load(positive_sample_path)
sad_sample = np.load(sad_sample_path)

positive_df = pd.DataFrame(positive_sample[0]).T
sad_df = pd.DataFrame(sad_sample[0]).T
test_df = pd.concat([positive_df, sad_df], axis=0)
labels = [0 for i in range(len(positive_df))]
labels += [1 for i in range(len(sad_df))]
scaled_test_data = min_max_scaler.fit_transform(test_df)
```

```python
som_ = som.SOM(9, 9)
som_.fit(scaled_test_data, 40000, save_e=True, verbose=False)
som_.plot_error_history()
som_.plot_point_map(scaled_test_data, labels, np.unique(labels))
som_.plot_distance_map()
```
