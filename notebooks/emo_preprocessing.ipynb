3. Построить гистограммы признаков
- Для построения гистограмм признаков собственных данных используйте соответствующие названия признаков при получении данных из датафрейма df['название_признака'].values

```python
import matplotlib.pyplot as plt
```

```python
def plot_features_hist(data, n_bins, x_count, y_count):
    y_count = 16
    x_count = 4
    for i in range(y_count):
        fig, axs = plt.subplots(1, x_count, figsize=(12, 4))
        for j in range(x_count):
            axs[j].hist(data[:, i * x_count + j], bins=n_bins)
            axs[j].set_title(f'{i * x_count + j}')

        plt.tight_layout()
        plt.show()
```

```python
def plot(data, labels, title, count_values_to_show=None):
    if count_values_to_show is None:
        count_values_to_show = len(data)
    plt.scatter(data[0:count_values_to_show, 0], data[0:count_values_to_show, 1], c=labels[:count_values_to_show])
    plt.title(title)
    plt.xlabel(0)
    plt.ylabel(1)
    plt.show()
```

```python
def plot_bars(data, emotions):
    get_values = lambda emotion : np.array([data[i] for i in range(len(data)) if emotions[i] == emotion])

    positive = get_values('positive')
    sad = get_values('sad')
    neutral = get_values('neutral')
    angry = get_values('angry')

    fig, axs = plt.subplots(2, 2, figsize=(10, 6))
    features = [i for i in range(0, 64)]
    axs[0, 0].bar(features, positive.mean(axis=0))
    axs[0, 0].set_title('positive')
    axs[0, 1].bar(features, sad.mean(axis=0))
    axs[0, 1].set_title('sad')
    axs[1, 0].bar(features, neutral.mean(axis=0))
    axs[1, 0].set_title('neutral')
    axs[1, 1].bar(features, angry.mean(axis=0))
    axs[1, 1].set_title('angry')

    plt.tight_layout()
    plt.show()
```

```python
data = df.to_numpy(dtype='float')
```

```python
plot_features_hist(data=data, n_bins=20, x_count=4, y_count=16)
```

```python
def transform_data_to_range(a, b, data):
    scaler = preprocessing.MinMaxScaler((a, b)).fit(data)
    scaled_data = scaler.transform(data)
    return scaled_data
```
