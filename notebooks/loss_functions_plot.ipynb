```python
import matplotlib.pyplot as plt
```

```python
def plot_train_val_loss(ax, train_loss, val_loss, title=''):
    epochs = [i + 1 for i in range(len(train_loss))]
    ax.plot(epochs, train_loss, label='Train Loss', linestyle='solid', color='g')
    ax.plot(epochs, val_loss, label='Val Loss', linestyle='dashed', color='b')
    
    ax.set_xlabel('epoch')
    ax.set_ylabel('loss')
    ax.set_title(title)
    
    ax.grid()
    ax.legend()
```

## 1000 audio

#### HuBERT clf

```python
fig, ax = plt.subplots(2, 2)
fig.set_figheight(6)
fig.set_figwidth(8)

plot_train_val_loss(ax[0, 0], [0.662, 0.526, 0.470], [0.595, 0.458, 0.419])
plot_train_val_loss(ax[0, 1], [0.662, 0.488, 0.422, 0.351], [0.558, 0.446, 0.434, 0.425])
plot_train_val_loss(ax[1, 0], [0.627, 0.449, 0.379, 0.304, 0.274], [0.571, 0.477, 0.574, 0.588, 0.646])
plot_train_val_loss(ax[1, 1], [0.636, 0.529, 0.416, 0.363, 0.300, 0.259], [0.599, 0.468, 0.487, 0.428, 0.435, 0.446])

plt.tight_layout()
plt.show()
```

#### HuBERT triplets

```python
fig, ax = plt.subplots(1, 3)
fig.set_figheight(4)
fig.set_figwidth(10)

plot_train_val_loss(ax[0], [0.979, 0.838, 0.715, 0.450, 0.295, 0.195], [0.900, 1.000, 0.592, 0.624, 0.461, 0.468], '1000 triplets')
plot_train_val_loss(ax[1], [0.961, 0.569, 0.335, 0.170], [0.801, 0.397, 0.143, 0.045], '2000 triplets')
plot_train_val_loss(ax[2], [0.928, 0.484, 0.182, 0.099], [0.642, 0.152, 0.106, 0.044], '3000 triplets')

plt.tight_layout()
plt.show()
```
